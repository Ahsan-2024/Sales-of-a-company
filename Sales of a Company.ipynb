Data Analysis

Gender

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender',data=df)# this line will make a plot chart
for bars in ax.containers:#these lines will count number of femals and males
    ax.bar_label(bars)#as we see females have purchased more than males
```

```python
sales_gen= df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)#1 ascending False will not make the output in desending order #2 sum function will sum the amount 
sns.barplot(x ='Gender', y= 'Amount', data=sales_gen)
```

```python
#so female have give more amount of purchasing 
```

Age

```python
df.columns
```

```python
ax = sns.countplot(data = df, x='Age Group', hue='Gender')
#hue will show how much are men and women
for bars in ax.containers:
    ax.bar_label(bars)#it will add count
```

```python
sales_age= df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x ='Age Group', y= 'Amount', data=sales_age)
```

```python
#so as we see 26-35 aage grp purchased more
```

State

```python
df.columns
```

```python
sales_State= df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})#it will give size to chart
sns.barplot(x ='State', y= 'Orders', data=sales_State)
```

```python
#more orders come from utar pardesh
```

```python
sales_State= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})#it will give size to chart
sns.barplot(x ='State', y= 'Amount', data=sales_State)
```

```python
#as we see our company generated more amount from utar pardesh 
```

MArtial Status

```python
df.columns
```

```python
ax = sns.countplot(data=df,x= 'Marital_Status')
for bars in ax.containers:
    sns.set(rc={'figure.figsize':(15,4)})
    ax.bar_label(bars)
```

```python
sales_State= df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(10,5)})#it will give size to chart
sns.barplot(x ='Marital_Status', y= 'Amount',hue='Gender', data=sales_State)
```

```python
#as we see married women have pirchased much
```

Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x= 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#as we see it sector order most
```

```python
sales_State= df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})#it will give size to chart
sns.barplot(x ='Occupation', y= 'Amount', data=sales_State)
```

```python
#we found top 3 #1 IT sector #2 Helath Care #3 Aviation
```

Product Category

```python
sns.set(rc={'figure.figsize':(25,5)})
ax =sns.countplot(data=df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#as we see top 3 sales product #1 Clothing #2Food #3 Electronics
```

```python
sales_State= df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(30,7)})#it will give size to chart
sns.barplot(x ='Product_Category', y= 'Amount', data=sales_State)
```

```python
#so now as we see that our most selled product is clothing but Food gave us more profit, so that is a important poimt
```

```python
sales_State= df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,7)})#it will give size to chart
sns.barplot(x ='Product_ID', y= 'Orders', data=sales_State)
```
